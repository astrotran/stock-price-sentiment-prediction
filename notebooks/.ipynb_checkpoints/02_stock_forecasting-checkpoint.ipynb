# Technical Indicators

## Exponential moving average

Tells you recent price emphasis vs. simple average

```python
df["EMA_30"] = df["Close"].ewm(span = 30, adjust = False).mean()

plt.figure(figsize = (12,6))
plt.plot(df["Close"], label = "Close", color = "blue")
plt.plot(df["EMA_30"], label = "30-day EMA", color = "orange")
plt.title("APPL Close Price and 30-day EMA")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

## Relative Strength Index (RSI)

Whether a stock is overbought or oversold

```python
#Calculate price change
delta = df["Close"].diff()

#Calculate gains and losses
gain = delta.clip(lower=0)
loss = -delta.clip(upper=0)

#Calculate average gain/loss
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()

#Calculate RS and RSI
rs = avg_gain / avg_loss
df["RSI"] = 100 - (100 / (1 + rs))

#Plot RSI
plt.figure(figsize=(12,4))
plt.plot(df["RSI"], label="RSI", color="purple")
plt.axhline(70, linestyle="--", alpha=0.5, color="red")
plt.axhline(30, linestyle="--", alpha=0.5, color="green")
plt.title("AAPL Relative Strength Index (RSI)")
plt.xlabel("Date")
plt.ylabel("RSI")
plt.legend()
plt.grid(True)
plt.show()
```

## Moving Average Convergence Divergence (MACD)

```python
#Short and long EMAs
ema_12 = df["Close"].ewm(span=12, adjust = False).mean()
ema_26 = df["Close"].ewm(span=26, adjust = False).mean()

#MACD and signal line
df["MACD"] = ema_12 - ema_26
df["MACD_signal"] = df["MACD"].ewm(span=9, adjust = False).mean()

#Plot
plt.figure(figsize=(12,6))
plt.plot(df["MACD"], label="MACD", color="blue")
plt.plot(df["MACD_signal"], label="Signal Line", color="red")
plt.axhline(0, color='black', linewidth=1, linestyle='--')
plt.title("AAPL MACD vs Signal Line")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

## Bollinger Bands

Price relative to volatility

```python
#20-day SMA and standard deviation
df["SMA_20"] = df["Close"].rolling(window=20).mean()
df["Upper_Band"] = df["SMA_20"] + 2 * df["Close"].rolling(window=20).std()
df["Lower_Band"] = df["SMA_20"] - 2 * df["Close"].rolling(window=20).std()

# Plot
plt.figure(figsize=(12,6))
plt.plot(df["Close"], label="Close", color="blue")
plt.plot(df["Upper_Band"], label="Upper Band", color="green", linestyle='--')
plt.plot(df["SMA_20"], label="20-day SMA", color="orange")
plt.plot(df["Lower_Band"], label="Lower Band", color="red", linestyle='--')
plt.title("AAPL Bollinger Bands")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

EMA: Price above 30-day EMA	Uptrend momentum

RSI:	         Falling but around mid-40s	Cooling off from overbought; neutral-ish

MACD:       	 Still above signal but converging	Trend still bullish, but possibly weakening

Bollinger Bands:	     Near upper band	Strong upward momentum, possible overbought


# Backtesting a Simple Strategy

Now that we have calculated trading signals such as EMA, RSI, and MACD, we can test if the signals actually make money. This is relevant for event-based trading logic, tracking cumulative return, and signal validation. We'll simulate buying when certain technical signals align, and track things like returns, hit rate, max drawdown, etc. and gain immediate feedback on whether these indicators work on a specific ticker.

```python
#prep the df
df = df.dropna(subset = ["EMA_30", "RSI", "MACD", "MACD_signal"]).copy()
```

```python
# BUY if Close > EMA and MACD > Signal Line (remove RSI)
#Trading condition
df["Signal"] = 0

df.loc[
    (df["Close"] > df["EMA_30"]) & (df["MACD"] > df["MACD_signal"]),
    "Signal"
] = 1

# SELL if Close < EMA or MACD < Signal Line
df.loc[
    (df["Close"] < df["EMA_30"]) | (df["MACD"] < df["MACD_signal"]),
    "Signal"
] = -1
```

```python
#Simulate Portfolio Performance
initial_cash = 10000
cash = initial_cash
shares = 0
portfolio_value = []

for i in range(len(df)):
    price = df["Close"].iloc[i]
    signal = df["Signal"].iloc[i]

    if signal == 1 and cash >= price:
        shares = cash // price # buy whole shares
        cash -= shares * price

    elif signal == -1 and shares > 0:
        cash += shares * price
        shares = 0

    value = cash + shares * price
    portfolio_value.append(value)

#Store portfolio value over time
df["Portfolio_Value"] = portfolio_value
```

```python
plt.figure(figsize=(14,6))
plt.plot(df.index, df["Portfolio_Value"], label="Strategy Portfolio Value", color="green")
plt.title("Portfolio Value Over Time (Simple Strategy)")
plt.xlabel("Date")
plt.ylabel("Portfolio Value ($)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.plot(df["Close"], label="Close Price", alpha=0.5)
plt.plot(df["EMA_30"], label="30-Day EMA", alpha=0.75)

# Plot buy signals
plt.plot(df[df["Signal"] == 1].index, df[df["Signal"] == 1]["Close"], "^", markersize=7 ,color='g', label='Buy Signal')

# Plot sell signals
plt.plot(df[df["Signal"] == -1].index, df[df["Signal"] == -1]["Close"], "v", markersize=7, color='r', label='Sell Signal')

plt.title("Buy/Sell Signals Based on Technical Indicators")
plt.xlabel("Date")
plt.ylabel("Price ($)")
plt.grid(True)
plt.legend()
plt.show()
```
