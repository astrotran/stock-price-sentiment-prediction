```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob

os.chdir("C:/Users/andre/projects/stock-price-sentiment-prediction/data/")
#path = "GOOGL_stock_prices.csv"

#ticker = 'GOOGL'

def main_function(file_path, ticker_name):
    import matplotlib.pyplot as plt
    df = pd.read_csv(file_path, skiprows = 3)
    df.columns = ["Date", "Close", "High", "Low", "Open", "Volume"]
    print(df.head())
    print("\nMissing values:\n", df.isnull().sum())
    print("\nData summary:\n", df.describe())
    
    #Handling missing values
    
    df_cleaned = df.dropna()
    
    df_cleaned = df.fillna(method = 'ffill')
    
    print("\nMissing values after cleaning:\n", df_cleaned.isnull().sum())
    
    df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])
    df_cleaned = df_cleaned.sort_values(by='Date')
    
    df_cleaned.set_index('Date', inplace = True)
    #Plot stock price trends
    plt.figure(figsize=(12,6))
    plt.plot(df_cleaned.index, df_cleaned['Close'], label = 'Close Price', color = 'blue')
    plt.title("Stock Price Trend - " + str(ticker_name))
    plt.xlabel("Date")
    plt.ylabel("Price (USD)")
    plt.legend()
    plt.show()
    
    
    df_cleaned['Daily Return'] = df_cleaned['Close'].pct_change()
    
    plt.figure(figsize=(12,6))
    plt.plot(df_cleaned.index, df_cleaned['Daily Return'], label = "Daily Returns", color = 'red')
    plt.axhline(0, linestyle = '--', color = 'black', alpha = 0.6)
    plt.title("Daily Returns - " + str(ticker_name))
    plt.xlabel("Date")
    plt.ylabel("Return")
    
    plt.legend()
    plt.show()
    
    print("\nVolatility (Std Dev of returns):", df_cleaned['Daily Return'].std())
    
    df_cleaned.to_csv(ticker_name + "_stock_prices_cleaned.csv")
    
    
#main_function(path,stock, ticker)
stock_directory = glob.glob("C:/Users/andre/projects/stock-price-sentiment-prediction/data/*_stock_prices.csv")

for file_path in stock_directory:
    path = file_path[62:]
    ticker = file_path[62:-17]

    main_function(path, ticker)
    print(path)

    
    
```
